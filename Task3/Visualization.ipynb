```python
import pandas as pd
data = pd.read_csv('Student_Marks.csv')
type(data)
pd.core.frame.DataFrame
data.info
```

```python
import matplotlib.pyplot as plt
import numpy as np
X = data["time_study"]
Y = data["Marks"]
plt.scatter(X,Y)
plt.xlabel('Study Hours')
plt.ylabel('Marks Obtained')
plt.legend()

plt.show()
```

```python
X = data["time_study"]
Y = data["Marks"]

plt.plot(X, Y, marker='o', color='blue')

plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Line Chart Example')

plt.legend()

plt.show()
```

```python
X = data["time_study"]
Y = data["Marks"]


plt.bar(X, Y, color='yellow')

plt.xlabel('Categories')
plt.ylabel('Values')
plt.title('Bar Chart Example')

plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
X = data["number_courses"]
Y = data["Marks"]
plt.scatter(X,Y)
plt.xlabel('Number of Courses')
plt.ylabel('Marks Obtained')
plt.legend()

plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
X = data["number_courses"]
Y = data["Marks"]
plt.bar(X,Y, color="green")
plt.xlabel('Number of Courses')
plt.ylabel('Marks Obtained')
plt.legend()

plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
from sklearn import preprocessing

data_cleaned = data.iloc[0:10]
min_max_scaler = preprocessing.MinMaxScaler()
data_normalized = pd.DataFrame(min_max_scaler.fit_transform(data_cleaned.values), columns=data_cleaned.columns)
sns.heatmap(data_normalized, annot=True)

plt.show()
```
